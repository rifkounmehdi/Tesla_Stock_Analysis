Tesla Stock Analysis Project
This notebook analyzes Tesla's historical stock data, computes daily returns, plots moving averages, and calculates basic statistics.

**Steps:**
1. Load the dataset (`Tesla_stock_price.csv`)
2. Calculate daily returns
3. Add a 50-day Simple Moving Average (SMA)
4. Plot stock price with SMA
5. Plot daily returns
6. Compute average return and volatility


```python

import pandas as pd
import matplotlib.pyplot as plt

# 1. Load dataset
data = pd.read_csv("Tesla_stock_price.csv")

# 2. Inspect first rows
data.head()

print(data.columns.tolist())
```

```python

# Convert Date column to datetime and set as index
data["Date"] = pd.to_datetime(data["Date"])
data.set_index("Date", inplace=True)


# 3. Calculate daily returns

# 3. Calculate daily returns
data["Return"] = data["Close"].pct_change()

# 4. Add 50-day SMA
data["SMA50"] = data["Close"].rolling(window=50).mean()

data.tail()
```

```python

# 5. Plot Tesla stock price with SMA50
plt.figure(figsize=(12,6))
plt.plot(data["Close"], label="Tesla Adjusted Close")
plt.plot(data["SMA50"], label="SMA 50 days", linestyle="--")
plt.title("Tesla Stock Price with 50-day SMA")
plt.legend()
plt.show()
```

```python

# 6. Plot daily returns
plt.figure(figsize=(12,6))
plt.plot(data["Return"], label="Daily Returns")
plt.axhline(0, color="red", linestyle="--")
plt.title("Tesla Daily Returns")
plt.legend()
plt.show()
```

```python

# 7. Basic statistics
avg_return = data["Return"].mean()
volatility = data["Return"].std()

print("Average daily return:", avg_return)
print("Volatility (std dev of returns):", volatility)
```
